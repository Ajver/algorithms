```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
```

```python
pre = pd.DataFrame(
    [
        [0.080, 0.076, 0.086],  # CEL
        [0.115, 0.113, 0.124],  # JOD
        [0.050, 0.053, 0.051],  # GAZ
    ],
    index=["CEL", "JOD", "GAZ"],
    columns=["1", "2", "3"],
)
pre
```

```python
post = pd.DataFrame(
    [
        [0.644, 0.667, 0.784],  # CEL
        [0.093, 0.109, 0.105],  # JOD
        [0.433, 0.408, 0.418],  # GAZ
    ],
    index=["CEL", "JOD", "GAZ"],
    columns=["1", "2", "3"],
)
post
```

```python
nasiak = 100*(post - pre)/pre  # %
nasiak
```

```python
mean_std = pd.concat([nasiak.mean(axis=1), nasiak.std(axis=1)], axis=1)
mean_std.columns = ["mean", "std"]
mean_std
```

```python
df = pd.concat([nasiak, mean_std], axis=1)
df
```

```python
print(df.to_latex())
```

```python
ph = pd.DataFrame(
    [
        [6.637, 6.669, 6.657],
        [6.787, 6.828, 6.740],
        [6.751, 6.767, 6.744],
    ],
    columns=["1", "2", "3"],
    index=["CEL", "JOD", "GAZ"],
)
ph
```

```python
ref_ph = 6.190
ref_ph
```

```python
mean_std = pd.concat([ph.mean(axis=1), ph.std(axis=1)], axis=1)
mean_std.columns = ["mean", "std"]
mean_std = mean_std.round(3)
mean_std
```

```python
diff = ph - ref_ph

diff_mean_std = pd.concat([diff.mean(axis=1), diff.std(axis=1)], axis=1)
diff_mean_std.columns=["śr. zmiana", "STD"]
diff_mean_std = diff_mean_std.round(3)
diff_mean_std
```

```python
t_stat, p_value = stats.ttest_rel(ph.to_numpy(), np.ones((3, 3)) * ref_ph)
t_stat, p_value
```

```python
ph_mean_std = pd.concat([ph, mean_std, diff_mean_std], axis=1)
ph_mean_std
```

```python
print(ph_mean_std.to_latex(float_format="%.3f"))
```
