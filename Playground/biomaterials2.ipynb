```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pre = pd.DataFrame({
    "1": [.080, .115, .050],
    "2": [.076, .113, .124],
    "3": [.086, .124, .051],
}, columns=["1", "2", "3"], index=["CEL", "JOD", "GAZ"])
pre
```

```python
post = pd.DataFrame({
    "1": [.644, .093, .433],
    "2": [.067, .109, .408],
    "3": [.784, .105, .418],
}, columns=["1", "2", "3"], index=["CEL", "JOD", "GAZ"])
post
```

```python
nasiak = 100*(post - pre)/pre  # %
nasiak
```

```python
df = pd.concat([nasiak.mean(axis=1), nasiak.std(axis=1)], axis=1)
df.columns = ["mean", "std"]
df
```
