```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pre = pd.DataFrame(
    [
        [0.080, 0.076, 0.086],  # CEL
        [0.115, 0.113, 0.124],  # JOD
        [0.050, 0.053, 0.051],  # GAZ
    ],
    index=["CEL", "JOD", "GAZ"],
    columns=["1", "2", "3"],
)
pre
```

```python
post = pd.DataFrame(
    [
        [0.644, 0.667, 0.784],  # CEL
        [0.093, 0.109, 0.105],  # JOD
        [0.433, 0.408, 0.418],  # GAZ
    ],
    index=["CEL", "JOD", "GAZ"],
    columns=["1", "2", "3"],
)
post
```

```python
nasiak = 100*(post - pre)/pre  # %
nasiak
```

```python
mean_std = pd.concat([nasiak.mean(axis=1), nasiak.std(axis=1)], axis=1)
mean_std.columns = ["mean", "std"]
mean_std
```

```python
df = pd.concat([nasiak, mean_std], axis=1)
df
```
