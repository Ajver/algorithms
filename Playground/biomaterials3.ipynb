```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
pre = pd.DataFrame(
    [
        [25, 26, 26],
        [18, 19, 19],
    ],
    columns=["1", "2", "3"],
    index=["AIR", "FREEZE"],
)*0.001
pre
```

```python
print(pre.to_latex(float_format="%.3f"))
```

```python
post = pd.DataFrame(
    [
        [1177, 1183, 1208],
        [644, 686, 695],
    ],
    columns=["1", "2", "3"],
    index=["AIR", "FREEZE"],
)*0.001
post
```

```python
print(post.to_latex(float_format="%.3f"))
```

```python
nasiak = 100*(post - pre)/pre  # %
nasiak
```

```python
mean_std = pd.concat([nasiak.mean(axis=1), nasiak.std(axis=1)], axis=1)
mean_std.columns = ["mean", "std"]
mean_std
```

```python
df = pd.concat([nasiak, mean_std], axis=1)
df
```

```python
print(df.to_latex(float_format="%.1f"))
df
```

```python
ph = pd.DataFrame(
    [
        [6.600, 6.581, 6.680],
        [6.700, 6.742, 6.667],
    ],
    columns=["1", "2", "3"],
    index=["AIR", "FREEZE"],
)
ph
```

```python
ref_ph = 6.190
ref_ph
```

```python
mean_std = pd.concat([ph.mean(axis=1), ph.std(axis=1)], axis=1)
mean_std.columns = ["mean", "std"]
mean_std = mean_std.round(3)
mean_std
```

```python
diff = ph - ref_ph

diff_mean_std = pd.concat([diff.mean(axis=1), diff.std(axis=1)], axis=1)
diff_mean_std.columns=["śr. zmiana", "STD"]
diff_mean_std = diff_mean_std.round(3)
diff_mean_std
```

```python
ph.to_numpy()[0]
```

```python
t_stat, p_value = stats.ttest_rel(ph.to_numpy(), np.ones((2, 3)) * ref_ph, axis=1)
p_value
```

```python
ph_mean_std = pd.concat([ph, mean_std, diff_mean_std], axis=1)
ph_mean_std
```

```python
print(ph_mean_std.to_latex(float_format="%.3f"))
```
