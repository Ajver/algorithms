```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr = np.array([6.20, 4.34, 8.14, 6.24, 3.72, 3.54, 4.35, 2.67, 7.16, 6.00])
plt.hist(arr, bins=10)
```

```python
mean = arr.mean()
mean
```

```python
std = arr.std()
std
```

```python
new_set = np.random.choice(arr, size=(1000, len(arr)), replace=True)
means_d = new_set.mean(axis=1)
```

```python
means_d_95 = np.sort(means_d)[25:975]
min_mean = means_d_95[0]
max_mean = means_d_95[-1]

print(f"95% Confidence Intervals: [{min_mean:.2f}, {max_mean:.2f}]")
```

```python
plt.hist(means_d, bins=20)
ylim = plt.ylim()
plt.axline((mean, ylim[0]), (mean, ylim[1]), c="k", linestyle="--")
plt.axline((min_mean, ylim[0]), (min_mean, ylim[1]), c="orange", linestyle="--")
plt.axline((max_mean, ylim[0]), (max_mean, ylim[1]), c="orange", linestyle="--")
```
